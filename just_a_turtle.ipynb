```python
from ColabTurtlePlus.Turtle import *

clearscreen()
setup(800, 600)
speed(10)
```

```python
def draw_petal(radius, angle):
    for _ in range(2):
        circle(radius, angle)
        left(180 - angle)

def draw_flower(petal_count, radius, color_name):
    color(color_name)

    for _ in range(petal_count):
        draw_petal(radius, 60)
        left(360 / petal_count)


    color("gold")
    penup()
    goto(xcor(), ycor())
    pendown()
    circle(10)


penup()
goto(0, -50)
pendown()

setheading(0)
draw_flower(petal_count=12, radius=80, color_name="magenta")

color("green")
pensize(4)
right(90)
penup()
forward(20)
pendown()
forward(180)

def draw_leaf():
    color("darkgreen")
    begin_fill()
    circle(30, 60)
    left(120)
    circle(30, 60)
    end_fill()

penup()
goto(0, -100)
setheading(200)
pendown()
draw_leaf()

penup()
goto(0, -140)
setheading(340)
pendown()
draw_leaf()
```
